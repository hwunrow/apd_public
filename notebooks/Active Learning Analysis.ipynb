```python
%matplotlib inline
import matplotlib.pylab as plt
import os, sys
import numpy as np
import pickle as pkl
from docopt import docopt
from itertools import product
```

```python
def _best(m):
    r = np.zeros(len(m))
    for i in range(len(m)):
        r[i] = np.max(m[:i+1])
    return r
```

```python
fig, ax = plt.subplots(figsize=[5.2, 3.4], dpi=130)
plt.style.use('ggplot')
t = np.load(os.path.join('../logs', 'BCentropy-apd:cnn-globe-bn-X-sgld-mnist-2-X-mnist@2018-05-28', 'acq_acc.npy'))
acq_arrs = np.array(t)
m = acq_arrs
plt.semilogx(np.arange(len(m))+1, _best(m),'.-', c='C3', label='Ours,Active')
t = np.load(os.path.join('../logs', 'BCentropy-cnn-globe-bn-X-sgld-mnist-2-X-mnist@2018-05-27', 'acq_acc.npy'))
acq_arrs = np.array(t)
m = acq_arrs
plt.semilogx(np.arange(len(m))+1, _best(m),'.-', c='C4', label='SGLD,Active')
t = np.load(os.path.join('../logs', 'DCentropy-mcd:cnn-globe-bn-X-sgld-mnist-2-X-mnist@2018-05-28', 'acq_acc.npy'))
acq_arrs = np.array(t)
m = acq_arrs
plt.semilogx(np.arange(len(m))+1, _best(m),'.-', c='C5', label='MC-Drop,Active')
plt.xlabel('num acquisition (T)')
plt.ylabel('best test acc up to T (/1.)')
plt.legend(loc='lower right')
plt.ylim([.3,1])

# plt.savefig('AL-cs.png', bbox_inches='tight')
```
